```python
def attrition_yes(df):
    return df[df['Attrition'] == 'Yes']

attrition_yes(df).head()
```

```python
def high_salary(df):
    return df[df['MonthlyIncome'] > 5000]
high_salary(df).head()
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="darkgrid")
```

```python
#Attribution count
sns.countplot(x='Attrition', data=df)
plt.title("Employee Attrition Count")
plt.show()
```

```python
#Department_wise_Employee
sns.countplot(x='Department', data=df)
plt.title("Employees by Department")
plt.xticks(rotation=30)
plt.show()
```

```python
#Salary Disbribution
sns.histplot(df['MonthlyIncome'], bins=30)
plt.title("Salary Distribution")
plt.show()
```

```python
#Attribution Vs Salary
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Attrition vs Salary")
plt.show()
```

```python
#job Satisfaction vs Attribution
sns.countplot(x='JobSatisfaction', hue='Attrition', data=df)
plt.title("Job Satisfaction vs Attrition")
plt.show()
```

```python
#Age Distribution
sns.histplot(df['Age'], bins=20)
plt.title("Age Distribution")
plt.show()
```

```python
#Avg Salary by Department
df.groupby('Department')['MonthlyIncome'].mean().plot(kind='bar')
plt.title("Avg Salary by Department")
plt.show()
```

```python
#Overtime Distribution
overtime_counts = df['OverTime'].value_counts()
plt.pie(overtime_counts, labels=overtime_counts.index, autopct='%1.1f%%')
centre_circle = plt.Circle((0,0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)
plt.title("Overtime Distribution")
plt.show()
```

```python
#Work-Life Balance Distribution
import matplotlib.pyplot as plt
wlb_counts = df['WorkLifeBalance'].value_counts()
plt.pie(wlb_counts, labels=wlb_counts.index, autopct='%1.1f%%')
plt.title("Work-Life Balance Distribution")
plt.show()
```

```python
#Employee Experience Level Distribution
exp_counts = df['ExperienceLevel'].value_counts()
plt.pie(exp_counts, labels=exp_counts.index, autopct='%1.1f%%')
centre_circle = plt.Circle((0,0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)
plt.title("Employee Experience Level Distribution")
plt.show()
```
